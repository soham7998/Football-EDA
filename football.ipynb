# IMPORT LIBRARIES 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('EPL_20_21.csv')
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
# Check for missing values
print(df.isnull().sum())

# Remove duplicates if any
df.drop_duplicates(inplace=True)
```

```python
df.describe()
```

```python
df.columns 
```

```python
df['MinsPerMatch'] = (df['Mins'] / df['Matches']).astype(int)
df['GoalsPerMatch'] = (df['Goals'] / df['Matches']).astype(float)
df['AssistsPerMatch'] = (df['Assists'] /df['Matches']).astype(float)

df.head()
```

```python
#Total amount of Goals scored in the 2020/21 season
Total_Goals = df['Goals'].sum()
Total_Goals
```

```python
#Total amount of Assists made in the 2020/21 season
Total_Assists = df['Assists'].sum()
Total_Assists
```

```python
#Total amount of penalty goals scored in the 2020/21 season
Total_PenaltyGoals = df['Penalty_Goals'].sum()
Total_PenaltyGoals
```

```python
#Total amount of Penaltys attempted in the 2020/21 season
Total_PenaltyAttempts = df['Penalty_Attempted'].sum()
Total_PenaltyAttempts
```

```python
#Pie Chart for penalties Missed vs Scored
plt.figure(figsize=(14,8))
py_not_scored = df['Penalty_Attempted'].sum() - Total_PenaltyGoals
py_data = [py_not_scored, Total_PenaltyGoals]
labels = ['Penalties missed', 'Penalties Scored']
color = sns.color_palette('Set2')
plt.pie(py_data, labels = labels, colors=color, autopct= '%.0f%%')
plt.show()
```

```python
#Most popular positions in the english premier league season 2020/21
pop_position = df.groupby('Position').size().sort_values(ascending=False)

pop_position.head(10).plot(kind='bar', figsize=(12, 6), color=sns.color_palette('Paired'))
plt.title("Most Popular positions in the EPL 2020/21 season", fontsize=18)
```

```python
pop_nation = df.groupby('Nationality').size().sort_values(ascending=False)

pop_nation.head(10).plot(kind='bar', figsize=(12,6), color=sns.color_palette("pastel"))
plt.title("Most Popular nationalities in the EPL 2020/21 season", fontsize=18)
```

```python
#Clubs with maximum players in their squad
df['Club'].value_counts().nlargest(5).plot(kind='bar', color=sns.color_palette('crest'))
plt.title("Clubs with the most of amount of Players in their squad", fontsize=20)
```

```python
#Players based on the age group
under20 = df.loc[df['Age'] <= 20]
age20_25 = df[(df['Age'] > 20) & (df['Age'] <= 25)]
age25_30 = df[(df['Age'] > 25) & (df['Age'] <= 30)]
above30  = df[df['Age'] > 30]
```

```python
x = np.array([under20['Name'].count(), age20_25['Name'].count(), age25_30['Name'].count(), above30['Name'].count()])
myLabels = ["<=20", ">20 & <= 25", "> 25 & <=30", "> 30"]
plt.title('Total Players with Age Groups', fontsize=20)
plt.pie(x, labels=myLabels, autopct="%.1f%%")
plt.show()
```

```python
#Total under 20 players in each club
players_under_20 = df[df['Age'] < 20]
players_under_20['Club'].value_counts().plot(kind ='bar', color=sns.color_palette("cubehelix"), figsize=(18, 8))
plt.title("Clubs with the most players under 20 in the 2020/21 season", fontsize=24)
```

```python
#Top 10 Assists
top_10_assists = df[['Name', 'Club', 'Matches', 'Assists']].nlargest(n=10, columns = 'Assists')
top_10_assists
```

```python
#Most goals by Players
#Top 10 Assists
top_10_goals = df[['Name', 'Club', 'Matches', 'Goals']].nlargest(n=10, columns = 'Goals')
top_10_goals
```

```python
#Top 10 Goals Per match
top_10_goals_per_match = df[['Name', 'Club', 'Matches', 'GoalsPerMatch']].nlargest(n=10, columns='GoalsPerMatch')
top_10_goals_per_match
```

```python
#Top 10 assisters under the age of 20
top_10_U20_assists = df[['Name', 'Age', 'Club', 'Matches', 'Assists']].loc[df['Age'] < 20].nlargest(n=10, columns='Assists')
top_10_U20_assists
```

```python
#Top 10 players with most yellow cards
epl_yellow = df.sort_values(by='Yellow_Cards', ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.title("Players with the most Yellow Cards in 2020/21 season", fontsize=26)
c = sns.barplot(x=epl_yellow['Name'], y=epl_yellow['Yellow_Cards'], label='Players', color='yellow')
plt.ylabel('Number of Yellow Cards')
c.set_xticklabels(c.get_xticklabels(), rotation=45)
```

```python
#Top 10 players with most yellow cards
epl_red = df.sort_values(by='Red_Cards', ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.title("Players with the most Red Cards in 2020/21 season", fontsize=26)
r = sns.barplot(x=epl_red['Name'], y=epl_red['Red_Cards'], label='Players', color='red')
plt.ylabel('Number of Red Cards')
r.set_xticklabels(r.get_xticklabels(), rotation=45)
```
